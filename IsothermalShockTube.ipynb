# Isothermal Shock Tube

* Solve the Sod shock tube problem in 1D with Lax-Friedrichs (very dissipative, but stable)

Ben Wibking - wibking.1@astronomy.ohio-state.edu

```python
import numpy as np

nx = 256
Lbox = 4.0
dx = Lbox/(nx+1)
x = np.linspace(-Lbox/2.,Lbox/2.,nx)

sigma = .5 # CFL condition
c = 1.0 # isothermal sound speed
dt = sigma*dx
t_final = 1.0
nt = int(t_final/dt)

u = np.zeros(nx,dtype=np.float64) # velocity
rho = np.ones(nx,dtype=np.float64) # density

rho[:nx//2] *= 3. # Sod shock tube problem

def evolve(rho,u):
    fac = c*c
    
    rho_array = np.zeros((nt+1,nx),dtype=np.float64)
    rho_array[0,:] = rho
    
    u_array = np.zeros((nt+1,nx),dtype=np.float64)
    u_array[0,:] = u
    
    for n in range(nt): ##loop across number of time steps
        rho_n = rho_array[n,:]
        u_n = u_array[n,:]
        
        momen = rho_n * u_n
        e_ext = 0.5 * rho_n * u_n * u_n + fac * rho_n
        
        ## Lax-Friedrichs method
        # mass conservation equation
        rho_array[n+1,1:-1] = 0.5*(rho_n[2:]+rho_n[0:-2]) - (dt/(2.*dx)*(momen[2:]-momen[0:-2]))
        
        # momentum conservation equation
        u_array[n+1,1:-1] = 0.5*(momen[2:]+momen[0:-2]) - (dt/(2.*dx)*(e_ext[2:]-e_ext[0:-2]))
        
        ## outflow boundary conditions
        rho_array[n+1,0] = (rho_n[0]) - (dt/(dx)*(momen[1]-momen[0]))
        rho_array[n+1,-1] = (rho_n[-1]) - (dt/(dx)*(momen[-1]-momen[-2]))
        
        u_array[n+1,0] = (momen[0]) - (dt/(dx)*(e_ext[1]-e_ext[0]))
        u_array[n+1,-1] = (momen[-1]) - (dt/(dx)*(e_ext[-1]-e_ext[-2]))

        u_array[n+1,:] /= rho_array[n+1,:] # save velocities only (but evolve in conserved variables)

    return rho_array,u_array
```

```python
plot(x,rho,label="density")
plot(x,u,label="velocity")
ylim(-0.5,3.5)
legend()
```

```python
density_array,vel_array = evolve(rho,u)
# in general, mass and momentum will not be conserved, because of outflow BCs
print "is mass conserved?",np.sum(density_array[0,:]),np.sum(density_array[-1,:])
print "is momentum conserved?",np.sum(density_array[0,:]*vel_array[0,:]),np.sum(density_array[-1,:]*vel_array[-1,:])
```
